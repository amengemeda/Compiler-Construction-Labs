## Lex

A Lex is a program generator designed for lexical processing of character input streams.

It accepts a high-level, problem oriented specification for character string matching, and produces a program in a general purpose language which recognizes regular expressions.

The regular expressions are specified by the user in the source specifications given to Lex and the Lex written code recognizes these expressions in an input stream and partitions the input stream into strings matching the expressions.

At the boundaries between strings program

The lexical analyzer creates a program in the Lex language and the Lex compiler runs the program and produces a C program.

The C compiler runs the C program and produces an object program which is the lexical analyzer that transforms an input stream into a sequence of tokens.

## Flex
Fast Lexical Analyzer Generator(Flex) is a tool for generating programs lexical analyzers, that perform pattern-matching on text.

It is used in writing compilers in conjuction with GNU Bison. It is a rewrite of the Unix Lex tool, although the two implementations do not share any code. Flex is about two times faster than Lex. It is also more flexible.

##*Bison*

Bison is a general-purpose generator that converts an annotated context-free grammar into a deterministic Left Right(LR) parser or generalized LR parser applying Look Ahead LR parser. Once one becomes proficient in Bison one can be able to develop language parser.

Bison is upward compatible with Yacc; properly written Yacc grammars are supposed to work with Bison

The Bison parser is a bottom-up parser. It tries, by shifts and reductions, to reduce the entire input down to a single grouping whose symbol is the grammar's start-symbol.

## Yacc

Yacc (Yet Another Compiler Compiler) is a tool that is used to generate a parser. It translates a given Context Free Grammar (CFG) specifications into a C implementation of a corresponding push down automaton (a finite state machine with a stack). <br><br>
Yacc is used to produce the source code of the syntactic analyzer of the language produced by LALR(1) (LookAhead, Left-to-right, Rightmost derivation producer with 1 lookahead token) grammar. <br><br>
The input of YACC is the rule or grammar, and the output is a C program.<br><br>
